```python
#Initial Imports
import os
import requests
import pandas as pd
import numpy as np
import datetime as dt
from pathlib import Path
from dotenv import load_dotenv
import alpaca_trade_api as tradeapi

%matplotlib inline

# Load .env environment variables
print(load_dotenv())
```

```python
# Current Amount of Shares 
ADP = 1000
BR = 500
CDK = 950
```

```python
# Set Alpaca API key and secret
alpaca_api_key = os.getenv("ALPACA_API_KEY")
alpaca_secret_key = os.getenv("ALPACA_SECRET_KEY")

# Create the Alpaca API object
api = tradeapi.REST(
    alpaca_api_key,
    alpaca_secret_key,
    api_version = "v2"
)
```

```python
# Format current date as ISO format
current_date = pd.Timestamp("2021-10-13", tz="America/New_York").isoformat()
```

```python
# Set the tickers
tickers = ["ADP", "BR", "CDK"]

# Set timeframe to '1D' for Alpaca API
timeframe = "1D"

# Get current closing prices for ADP, BR, and CDK
df_ticker = api.get_barset(
    tickers,
    timeframe,     limit=1000
).df

# Preview DataFrame
df_ticker

```

```python
# Pick ADP, BR and CDK close prices
ADP_close_price = df_ticker['ADP']['close'][1000]
BR_close_price = df_ticker['BR']['close'][1000]
CDK_close_price = df_ticker['CDK']['close'][1000]

# Print ADP, BR and CDK close prices
print(f"Current ADP closing price: $(adp_close_price)")
print(f"Current BR closing price: $(br_close_price)")
print(f"Current CDK closing price: $(cdk_close_price)")
```

```python
#Compute the current value of shares
adp_value = ADP_close_price * ADP
br_value = BR_close_price * BR
cdk_value = CDK_close_price * CDK
```

```python
# print current values of shares
print(f"The current value of {ADP} ADP shares is ${adp_value:0.2f}")
print(f"The current value of {BR} BR shares is ${br_value:0.2f}")
print(f"The current value of {CDK} CDK shares is ${cdk_value:0.2f}")
```

```python
# Calculate Anthon Portfolio Value
anthon_portfolio = adp_value + br_value + cdk_value
print('Anthon portfolio value:', anthon_portfolio)
```

```python
# Amount of Shares in Proposed Portfolio
AAPL = 500
GOOG = 48
NFLX = 100
NVDA = 320
```

```python
# Set the tickers
tickers = ["AAPL", "GOOG", "NFLX", "NVDA"]

# Set timeframe to '1D' for Alpaca API
timeframe = "1D"

# Get current closing prices for AAPL, GOOG, NFLX, NVDA
df_tickernew = api.get_barset(
    tickers,
    timeframe,
     limit=1000
).df

# Preview DataFrame
df_tickernew
```

```python
# Pick AAPL, GOOG, NFLX, and NVDA close prices
AAPL_close_price = df_tickernew['AAPL']['close'][-1]
GOOG_close_price = df_tickernew['GOOG']['close'][-1]
NFLX_close_price = df_tickernew['NFLX']['close'][-1]
NVDA_close_price = df_tickernew['NVDA']['close'][-1]

# Print AAPL, GOOG, NFLX, and NVDA close prices
print(f"Current AAPL closing price: ${AAPL_close_price}")
print(f"Current GOOG closing price: ${GOOG_close_price}")
print(f"Current NFLX closing price: ${NFLX_close_price}")
print(f"Current NVDA closing price: ${NVDA_close_price}")
```

```python
#Compute the current value of shares
aapl_value = AAPL_close_price * AAPL
goog_value = GOOG_close_price * GOOG
nflx_value = NFLX_close_price * NFLX
nvda_value = NVDA_close_price * NVDA
```

```python
# print current values of shares
print(f"The current value of {AAPL} AAPL shares is ${aapl_value:0.2f}")
print(f"The current value of {GOOG} GOOG shares is ${goog_value:0.2f}")
print(f"The current value of {NFLX} NFLX shares is ${nflx_value:0.2f}")
print(f"The current value of {NVDA} NVDA shares is ${nvda_value:0.2f}")

```

```python
# Calculate Proposed Portfolio Value
proposed_portfolio = aapl_value + goog_value + nflx_value +nvda_value
print(f'Proposed portfolio value: ${proposed_portfolio:0.2f}')
```
