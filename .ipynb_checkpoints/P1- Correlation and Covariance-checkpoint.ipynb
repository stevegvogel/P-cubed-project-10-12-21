# P. Anthon Stocks Returns - Correlation 

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import pandas_datareader as web
```

```python
tickers = ["ADP", "BR", "CDK"]
portfolio_stocks = web.get_data_yahoo(tickers,
start = "2017-10-12",
end = "2021-10-12")
```

```python
# Charts of stock prices 
fig = plt.figure()
ax1 = fig.add_subplot(321)
ax2 = fig.add_subplot(322)
ax3 = fig.add_subplot(323)
ax1.plot(portfolio_stocks['Adj Close']['ADP'])
ax1.set_title("ADP")
ax2.plot(portfolio_stocks['Adj Close']['BR'])
ax2.set_title("BR")
ax3.plot(portfolio_stocks['Adj Close']['CDK'])
ax3.set_title("CDK")
plt.show()
```

```python
# Returns of stocks
portfolio_stock_daily_returns = portfolio_stocks['Adj Close'].pct_change()
portfolio_stock_monthly_returns = portfolio_stocks['Adj Close'].resample('M').ffill().pct_change()
```

```python
fig = plt.figure()
(portfolio_stock_monthly_returns + 1).cumprod().plot()
plt.show()
```

```python
# Calculate the mean and standard deviation of our portfolio
print(portfolio_stock_monthly_returns.mean())
```

```python
print(portfolio_stock_monthly_returns.std())
```

```python
# Calculate the correlation and covariance 
print(portfolio_stock_monthly_returns.corr())
```

```python
# Calculate the covariance 
print(portfolio_stock_monthly_returns.cov())
```
